```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
results = pd.read_csv("Content_Model_Testing_Results.csv")
```

```python
results
cosine = results.loc[results['Similarity Metric'] == 'Cosine Similarity']['RMSE']
pearson = results.loc[results['Similarity Metric'] == 'Pearson Correlation']['RMSE']
pearson
results
print cosine
print pearson
```

```python
num_of_weight_combinations=4
bar_width = .35
ind = np.arange(num_of_weight_combinations)
fig, ax = plt.subplots()
fig.set_size_inches(10,7)
plt.ylim(1, 2)
rects1 = ax.bar(ind, cosine, bar_width, color="orange")
rects2 = ax.bar(ind+bar_width, pearson, bar_width, color="blue")
line = ax.axhline(y=1.8419, color='r', linestyle='--')
ax.set_ylabel("RMSE")
ax.set_title("Content based algorithm with different weights to features and similarity mnetrics")
ax.set_xticks(ind + bar_width / 2)
#ax.set_yticks(np.arange(1.2, 1.3, 0.001))
#ax.set_ylim(bottom=.9, top=.95)
ax.set_xticklabels(('Equal weight', '2-1-1-2', '1-2-2-1', '1-5-5-1'))

ax.legend((rects1[0], rects2[0], line), ('Cosine Similarity', 'Pearson Similarity', 'Random'))
plt.show()
```
