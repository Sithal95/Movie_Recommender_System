```python
import pickle
from imdb import IMDb
import time

movies = {}
for l in open('ml-latest/links.csv').readlines()[1:]:
    x = l.strip().split(',')[1]
    movies[x] = 0
for l in open('ml-20m/links.csv').readlines()[1:]:
    x = l.strip().split(',')[1]
    movies[x] = 0
for l in open('ml-latest-small/links.csv').readlines()[1:]:
    x = l.strip().split(',')[1]
    movies[x] = 0
pickle.dump(movies,open('data/imdb_movie_ids.dat','wb'))

print len(movies)
```

```python
ia = IMDb()
c = 0
data = {}
id_file = {}

for x in movies:
    temp = ia.get_movie(x)
    data[x] = temp
    c+=1
    id_file[x] = 1+(c-1)/1000
    print x,c,time.time()
    if c%1000==0:
        pickle.dump(data,open('data/data'+str(c/1000)+'.dat','wb'))
        data = {}
pickle.dump(data,open('data/data'+str((c/1000)+1)+'.dat','wb'))
data = {}
pickle.dump(id_file,open('data/id_file.dat','wb'))
```
