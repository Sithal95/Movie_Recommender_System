```python
import numpy
import pickle
import random

data = open('./new_data.csv')
data = data.readlines()
print len(data)
c = 0
user_data = {}
users = {}
items = {}
for line in data:
    c+=1
    user,item,rating = line.strip().split(',')
    if int(user) not in user_data:
        user_data[int(user)] = []
    users[int(user)] = 0
    items[int(item)] = 0
    user_data[int(user)].append((int(item),float(rating)))
    
    if c%100000==0:
        print c
        
print c,len(users),len(items),max(users.keys()),max(items.keys())
```

```python
f1 = open('./split_1.csv','w')
f2 = open('./split_2.csv','w')
f3 = open('./split_3.csv','w')
f4 = open('./split_4.csv','w')
f5 = open('./split_5.csv','w')
for user in user_data:
    random.shuffle(user_data[user])
    size = len(user_data[user])
    base = size/5
    c = 0
    n = 0
    s1 = ""
    s2 = ""
    s3 = ""
    s4 = ""
    s5 = ""
    for x in user_data[user]:
        item = x[0]
        rating = x[1]
        if n<base:
            s1+=str(user)+","+str(item)+","+str(rating)+"\n"
        elif n<2*base:
            s2+=str(user)+","+str(item)+","+str(rating)+"\n"
        elif n<3*base:
            s3+=str(user)+","+str(item)+","+str(rating)+"\n"
        elif n<4*base:
            s4+=str(user)+","+str(item)+","+str(rating)+"\n"
        elif n<5*base:
            s5+=str(user)+","+str(item)+","+str(rating)+"\n"
        else:
            c+=1
            if c==1:
                s1+=str(user)+","+str(item)+","+str(rating)+"\n"
            elif c==2:
                s2+=str(user)+","+str(item)+","+str(rating)+"\n"
            elif c==3:
                s3+=str(user)+","+str(item)+","+str(rating)+"\n"
            elif c==4:
                s4+=str(user)+","+str(item)+","+str(rating)+"\n"
        n+=1
    f1.write(s1)
    f2.write(s2)
    f3.write(s3)
    f4.write(s4)
    f5.write(s5)
f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
```
